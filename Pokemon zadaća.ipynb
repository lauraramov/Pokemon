```python
def Simple_battle(poke1,poke2):
    winner='Equally strong, no winners'
    total1=(pokemon.loc[poke1])['Total']
    total2=(pokemon.loc[poke2])['Total']
    
    if(total1<total2):
        winner=poke2
    if(total1>total2):
        winner=poke1
    return winner;
```
